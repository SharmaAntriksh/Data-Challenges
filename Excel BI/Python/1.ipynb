```python
import pandas as pd
import re
```

```python
df = pd.read_excel(fr"C:\Users\antsharma\Downloads\PQ_Challenge_1_Problem.xlsx", skiprows = 1, nrows=10,  usecols= 'B')
df
```

```python
def get_digits(text):
    return ''.join(re.findall(r'\d+', str(text)))
    
df['Digits'] = df['String'].apply(get_digits)
df['Sum of Digits'] = df['String'].apply(lambda x: sum([int(i) for i in get_digits(x)]))
df['Sum of Digits 2'] = df['Digits'].apply(lambda x: sum([int(i) for i in x]))
```
